```python
#Here we'll try to analyse a dataset containing montly unemployment rates for a couple of years. 
#The data is collected by the United States Bureau of Labor Statistics.
#The unemployment rates may not be perfect, but it's often used as a proxy for the health of the economy. 
```

```python
import pandas as pd
unrate = pd.read_csv('unrate.csv')
print('unrate.head(12)')
print(unrate.head(12))
```

```python
print('unrate.tail(12)')
print(unrate.tail(12))
```

```python
#Data is listed per month of each year from January 1948 to August 2016.
#We want to have a look at how the unemployment rate changes over the years.
#If we group the data by year values, and make seperate plots for each year, comparing so many plots will be very difficult.
#Instead we can consider only the month values from DATE column and overlay all linecharts for different years on the same x-axis. 
#For that at first we have to extract the month values from the dataset.
```

```python
unrate['DATE'] = pd.to_datetime(unrate['DATE'], dayfirst=True)
unrate['MONTH'] = unrate['DATE'].dt.month
print(unrate.head(12))
```

```python
#As times passes by, the unemployment rate between 2 data points fluctuate.
#So, line-charts will be good for visualising the trends in the dataset.
#As of now, let's plot the data for the first 5 years only.
#We will use different colours for each year for easy understanding.
```

```python
import matplotlib.pyplot as plt
fig = plt.figure(figsize=(10,6))
colors = ['red', 'blue', 'green', 'orange', 'black']
for i in range(5):
    start_index = i*12
    end_index = (i+1)*12
    subset = unrate[start_index:end_index]
    label = str(1948 + i)
    plt.plot(subset['MONTH'], subset['VALUE'], c=colors[i], label=label)
plt.legend(loc='upper left')
plt.xlabel('Month, Integer')
plt.ylabel('Unemployment Rate, Percent')
plt.title('Monthly Unemployment Trends, 1948-1952')
plt.show()
```
